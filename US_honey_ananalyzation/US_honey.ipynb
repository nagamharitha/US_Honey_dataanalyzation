# Analysing the inferences from the data

Which states are rarely contributing for honey production since last 27 years

```python
data = df['state'].value_counts()
```

```python
data.index
```

```python
data.values
```

```python
plt.figure(figsize = (13,13))
plt.pie(data.values,labels=data.index,autopct="%0.2f%%")
plt.show()
```

```python
#inference: top 5 states contributing less in last 27 years are:
# maryland  9
#oklamhoma 9
#south carolina = 12
#new mexico 18
#nevada = 15
```

```python
data[data < data.mean()]
```

which are the top 5 honey producing states

```python
df.sort_values('production', ascending = False).head(5)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=df,x="state",y="production")
plt.xticks(rotation=90)
plt.show()
```

```python
data = df.groupby('state').sum().reset_index()
```

```python
data = data.sort_values('production',ascending=False)
data
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=data,x="state",y="production")
plt.xticks(rotation=90)
plt.show()
```

```python
#inference
#top 5 honey producing states

#north dakota
#california
#south Dakota
##florida
#montana
```

change in mean Avearge price of honey from 1995 to 2021

```python
data = df.groupby('year').mean()['average_price'].reset_index()
```

```python
data
```

```python
plt.figure(figsize=(14,12))
sns.barplot(data=data,x='year',y='average_price')
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.plot(data['year'],data['average_price'])
plt.xticks(rotation=90)
plt.show()
```

```python
#inference

#from year 1995 to 2017 we can see a increasing trend in avg price og honey but after 2017 the avg price falls suddenly and remains
#constant till 2021
```

which was the year when the production of honey in whole US was the highest?

```python
data = df.groupby('year').sum()['production'].reset_index()
```

```python
data
```

```python
plt.figure(figsize = (12,8))
plt.plot(data['year'],data['production'])
plt.xticks(rotation=90)
plt.show()
```

```python
data1 = df.groupby('year').sum()['production'].reset_index()

plt.figure(figsize = (13,8))

sns.barplot(data = data1, x = 'year',y = 'production')
plt.xticks(rotation = 90)
plt.show()
```

```python
#inference :- production of honey decreasing drasctically

```

from the abouve plots we found that year 2000 has highest production , now let's check which state has the highest contribution
in it

```python
data1 = df[df['year']==2000]
```

```python
data1
```

```python
plt.figure(figsize=(13,8))
sns.barplot(data=data1,x='state',y='production')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(30,30))
plt.pie(data1['production'],labels=data1['state'],autopct='%0.2f%%')
plt.show()
```

```python
#in year 2000 north dakota has the highest involvemnt in honey production
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=data1,x='state',y='colonies_number')
plt.xticks(rotation=90)
plt.show()
```
